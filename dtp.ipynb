Заказчиком выступает проект «Карта ДТП» https://dtp-stat.ru/ — некоммерческий проект, посвященный проблеме дорожно-транспортных происшествий (ДТП) в России. Это платформа сбора данных о ДТП, бесплатный и открытый сервис аналитики ДТП.

**Цель проекта** - провести глубокий анализ данных, сформулировать и проверить гипотезы, опираясь на доступные признаки.

## Загрузка данных

```python
import pandas as pd
import json
from matplotlib import pyplot as plt
import matplotlib
import seaborn as sns
import plotly.express as px
import numpy as np
import math
from scipy import stats as st
from scipy.stats import chi2_contingency
from scipy.stats import linregress
from sklearn.linear_model import LinearRegression
```

```python

# Загрузка GeoJSON файла
file_paths = [
    "respublika-bashkortostan.geojson", 
    "moskva.geojson",
    "iaroslavskaia-oblast.geojson",
    "kaliningradskaia-oblast.geojson",
    "khabarovskii-krai.geojson",
    "khanty-mansiiskii-avtonomnyi-okrug-iugra.geojson",
    "sankt-peterburg.geojson",
    "chechenskaia-respublika.geojson",
    "krasnodarskii-krai.geojson"
] 
# Список для хранения всех строк данных
rows = []

# Обрабатываем каждый файл в списке
for file_path in file_paths:
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)

    # Проверяем структуру данных
    features = data.get("features", [])

    for feature in features:
        properties = feature.get("properties", {})
        geometry = feature.get("geometry", {})

        # Основные поля
        row = {
            "id": properties.get("id"),
            "tags": ", ".join(properties.get("tags", [])),
            "light": properties.get("light"),
            "lat": geometry.get("coordinates", [None, None])[1],
            "long": geometry.get("coordinates", [None, None])[0],
            "region": properties.get("region"),
            "address": properties.get("address"),
            "weather": ", ".join(properties.get("weather", [])),
            "category": properties.get("category"),
            "datetime": properties.get("datetime"),
            "severity": properties.get("severity"),
            "dead_count": properties.get("dead_count"),
            "injured_count": properties.get("injured_count"),
            "participants_count": properties.get("participants_count"),
            "road_conditions": ", ".join(properties.get("road_conditions", [])),
            "participant_categories": ", ".join(properties.get("participant_categories", [])),
            "nearby": ", ".join(properties.get("nearby", [])),
            "parent_region": properties.get("parent_region")
        }

        # Обработка vehicles
        vehicles = properties.get("vehicles", [])
        if vehicles:
            for key in ["year", "brand", "color", "model", "category"]:
                row[f"vehicle_{key}"] = ", ".join(
                    str(vehicle.get(key, "")) for vehicle in vehicles
                )

        # Обработка participants
        participants = properties.get("participants", [])
        if participants:
            for key in ["role", "gender", "violations", "health_status", "years_of_driving_experience"]:
                row[f"participant_{key}"] = ", ".join(
                    str(participant.get(key, "")) for participant in participants
                )

        rows.append(row)


# Преобразуем в DataFrame
df = pd.DataFrame(rows)

# Сохраняем в CSV
df.to_csv("processed_data.csv", index=False)

pd.set_option('display.max_columns', None)

# Просмотр первых строк
print(df.head())
```

```python
# Функция для группировки категорий
def categorize(row):
    if row.startswith('Наезд'):
        return 'Наезд'
    elif row == 'Столкновение':
        return 'Столкновение'
    elif row == 'Опрокидывание':
        return 'Опрокидывание'
    elif row == 'Съезд с дороги':
        return 'Съезд с дороги'
    elif row == 'Падение пассажира':
        return 'Падение пассажира'
    else:
        return 'Другое'

# Добавление нового столбца
df['category_grouped'] = df['category'].apply(categorize)

# Проверка результата
df[['category', 'category_grouped']].head()
```

Теперь сделаем этот график с разбивкой по регионам.

Чтобы построить график с разбивкой по регионам, нужно группировать данные по времени суток и регионам, а затем визуализировать данные для каждого региона. Используем столбчатую диаграмму с группировкой по регионам.


```python
# Подсчет количества аварий по времени суток и регионам
light_region_counts = df.pivot_table(
    index='light', 
    columns='parent_region', 
    values='id', 
    aggfunc='count', 
    fill_value=0
)

# Общая сумма для расчета процентов
region_totals = light_region_counts.sum(axis=0)
light_region_relative = light_region_counts.div(region_totals, axis=1) * 100

# Настройка размеров графика
plt.figure(figsize=(14, 8))

# Построение горизонтальной столбчатой диаграммы
light_region_relative.T.plot(
    kind='bar', 
    stacked=True, 
    figsize=(14, 8), 
    colormap='viridis', 
    width=0.8
)

# Добавление заголовков и меток
plt.title("Распределение аварий по времени суток и регионам", fontsize=16)
plt.xlabel("Регион", fontsize=12)
plt.ylabel("Процент от общего числа аварий", fontsize=12)
plt.xticks(fontsize=10, rotation=45, ha='right')
plt.yticks(fontsize=10)
plt.legend(
    title="Время суток", 
    bbox_to_anchor=(1.0, 0.5),  # Легенда расположена ближе к центру графика
    fontsize=10,                
    title_fontsize=12,          
    frameon=False               # Убираем рамку легенды
)

# Добавление сетки
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Отображение графика
plt.tight_layout()
plt.show()
```

**В разрезе регионов мы видим, что минимальное значение ДТП в темное время суток с отсутствующим освещением в Москве и Саннкт-Петербурге. СКорее всего это связано с инфраструктурой этих городов,т.к. там освещение максимально присутствует во всех частях региона. Судя по этому мы делаем вывод, что в Чеченской республике гораздо больше мест, где освещение отсутствует, поэтому там и больше ДТП в вечернее время. Сумерки занимают довольно маленькую часть на графиках, потому что они вцелом и в сутках длятся всего несколько часов, а в зимние период и того меньше.**

### Изменение смертности в ДТП по годам и регионам

```python
# Создаем столбец с годами из datetime
df['year'] = pd.to_datetime(df['datetime']).dt.year
```

```python
# Группируем данные по годам и регионам для общего числа ДТП
yearly_region_total = (
    df.groupby(['year', 'parent_region'])['id']  
    .count()
    .reset_index()
    .rename(columns={'id': 'total_accidents'})  # Переименовываем столбец
)
yearly_region_total.head()
```

```python
# Группируем данные по годам и регионам и суммируем количество смертей
yearly_region_deaths = (
    df.groupby(['year', 'parent_region'])['dead_count']
    .sum()
    .reset_index()
)
yearly_region_deaths.head()
```

```python
# Объединяем обе таблицы по годам и регионам
yearly_region_data = pd.merge(yearly_region_deaths, yearly_region_total, on=['year', 'parent_region'])
yearly_region_data.head()
```

Корреляция имеет смысл только для числовых переменных. 
В нашем случае это:

- lat (широта, float)
- long (долгота, float)
- dead_count (количество погибших, int64)
- injured_count (количество раненых, int64)
- participants_count (количество участников, int64)

Используем метод `.corr()`, чтобы построить матрицу корреляций между числовыми столбцами.

```python
# Выбираем только числовые столбцы
correlation_matrix = df[['lat', 'long', 'dead_count', 'injured_count', 'participants_count']].corr()

# Выводим матрицу корреляций
print(correlation_matrix)
```

```python
# Выбираем только числовые столбцы для корреляции
numeric_columns = ['lat', 'long', 'dead_count', 'injured_count', 'participants_count']
correlation_matrix = df[numeric_columns].corr()

# Перебираем все пары столбцов
pairs = [(col1, col2) for i, col1 in enumerate(numeric_columns) for j, col2 in enumerate(numeric_columns) if i < j]

# Расчет необходимого числа строк и столбцов для подграфиков
num_pairs = len(pairs)
num_cols = 3  # Количество колонок
num_rows = math.ceil(num_pairs / num_cols)  # Округляем в большую сторону

# Настройки для графика
plt.figure(figsize=(16, 12))

# Строим scatter plot для каждой пары
for i, (col1, col2) in enumerate(pairs, start=1):
    plt.subplot(num_rows, num_cols, i)  # Подклетка для каждого scatter plot
    plt.scatter(df[col1], df[col2], alpha=0.5)
    plt.title(f'{col1} vs {col2}')
    plt.xlabel(col1)
    plt.ylabel(col2)

plt.tight_layout()
plt.show()
```

**Количество участников (participants_count) сильно связано с количеством пострадавших, что логично, потому что чем больше участников в ДТП, тем больше шансов на травмы. Таким образом, можно сделать вывод, что географическое положение не является значимым фактором для смертности в ДТП, но количество участников может быть важным предиктором как для числа пострадавших, так и для серьезности аварии.**

## Проверка гипотез
### Гипотеза о влиянии мотоциклистов на степень тяжести ДТП

Чтобы проверить статистически значимую разницу в тяжести ДТП, можно использовать **критерий χ².**

Критерий χ² (хи-квадрат) используется для проверки статистической гипотезы о взаимосвязи двух категориальных переменных. В данном случае, мы пытаемся проверить, существует ли статистически значимая разница в тяжести ДТП для разных групп участников, например, мотоциклистов. Вот почему критерий χ² является хорошим выбором:

- обе переменные (тяжесть ДТП и категория участников) являются категориальными. Критерий χ² помогает проверить, связаны ли эти переменные между собой.
- набор данных достаточно велик, чтобы ожидать, что все ячейки таблицы сопряженности имеют более или менее значимые значения

Тест на независимость: Мы проверяем гипотезу о независимости двух категориальных переменных (т.е. есть ли связь между категорией участников ДТП и тяжестью этих ДТП). Если переменные независимы, то мы не ожидаем, что одна влияет на другую. Если они зависимы, то тяжесть ДТП для мотоциклистов будет отличаться от тяжести ДТП для других участников.

Ожидаемые частоты: Критерий χ² рассчитывает, насколько наблюдаемые частоты (реальные данные) отличаются от ожидаемых частот (если бы переменные были независимы). Этот критерий используется для анализа таблиц сопряженности.

**Нулевая гипотеза:** Наличие мотоциклистов в ДТП не связано с тяжестью ДТП.

**Альтернативная гипотеза:** Наличие мотоциклистов в ДТП связано с изменением тяжести ДТП.

```python
# Выделяем ДТП с мотоциклистами и без них
df['is_motorcyclist'] = df['participant_categories_grouped'].apply(lambda x: 'Мотоциклисты' in x)

# Группируем данные по наличию мотоциклистов и тяжести ДТП
severity_motorcyclist = df.groupby(['is_motorcyclist', 'severity'])['id'].count().unstack()
severity_motorcyclist
```

```python
# Процентное распределение тяжести
severity_motorcyclist_percentage = severity_motorcyclist.div(severity_motorcyclist.sum(axis=1), axis=0) * 100

# Вывод результата
severity_motorcyclist_percentage
```

```python
# Создаем таблицу сопряженности
contingency_table = severity_motorcyclist.fillna(0).values

# Применяем χ² тест
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Результат
print(f"Хи-квадрат: {chi2}, p-value: {p}")
if p < 0.05:
    print("Разница в тяжести ДТП между группами статистически значима.")
else:
    print("Разница в тяжести ДТП между группами не статистически значима.")
```

**Отклонение нулевой гипотезы: Различия в тяжести ДТП между ДТП с участием мотоциклистов и без них статистически значимы. Таким образом, тяжесть ДТП для мотоциклистов отличается от тяжести ДТП для других участников. Мотоциклисты влияют на тяжесть ДТП. Это может быть связано с их большей уязвимостью на дорогах, высокими скоростями, ограниченной защитой при авариях и характером их участия в ДТП.**

### Гипотеза: "Количество аварий зависит от месяца года."

Для проверки гипотезы о том, что количество аварий зависит от месяца года, анализ дисперсии (ANOVA) будет наиболее подходящим методом. ANOVA позволяет сравнивать средние значения нескольких групп, в данном случае, группы — это месяцы года, и проверяет, существует ли статистически значимая разница в количестве аварий между этими группами.

**Нулевая гипотеза:** Среднее количество аварий в каждом месяце одинаково (нет зависимости от месяца).

**Альтернативная гипотеза:** Среднее количество аварий в разных месяцах отличается (есть зависимость от месяца).


```python
# Добавляем новый столбец 'year' и 'month' на основе 'datetime'
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month

# Группируем данные по году и месяцу и считаем количество аварий
accidents_per_month_year = df.groupby(['year', 'month']).size().reset_index(name='accidents_count')

accidents_per_month_year
```

```python
# Разделим данные по месяцам (для каждого месяца, например, все аварии по январям, февралям и т.д.)
# Каждая группа будет представлять аварии за месяц по разным годам
accidents_grouped = [accidents_per_month_year[accidents_per_month_year['month'] == month]['accidents_count']
                     for month in range(1, 13)]

# Проведем ANOVA тест для всех месяцев
f_stat, p_value_anova = st.f_oneway(*accidents_grouped)

# Выводим результаты
print(f"F-статистика: {f_stat}")
print(f"p-значение: {p_value_anova}")

# Проверяем гипотезу
if p_value_anova < 0.05:
    print("Отвергаем нулевую гипотезу: существует статистически значимая разница в количестве аварий между месяцами.")
else:
    print("Не отвергаем нулевую гипотезу: нет статистически значимой разницы в количестве аварий между месяцами.")
```

Мы убедились, что существует статистически значимая разница в количестве аварий между месяцами. Предположим, что в сентябре происходит больше всего аварий. 

Для проверки гипотезы можно использовать t-тест для проверки статистической значимости различий между сентябрем и другими месяцами. 

В данном случае мы проверим гипотезу:

**Нулевая гипотеза:** Количество аварий в сентябре не отличается от количества аварий в других месяцах.

**Альтернативная гипотеза:** Количество аварий в сентябре больше, чем в других месяцах.


```python
# Извлекаем данные о сентябре (месяц = 9) и другие месяцы
september_accidents = accidents_per_month_year[accidents_per_month_year['month'] == 9]['accidents_count']
other_months_accidents = accidents_per_month_year[accidents_per_month_year['month'] != 9]['accidents_count']

# Проведем t-тест для независимых выборок
t_stat, p_value_ttest = st.ttest_ind(september_accidents, other_months_accidents)

# Выводим результаты
print(f"T-статистика: {t_stat}")
print(f"p-значение: {p_value_ttest}")

# Проверяем гипотезу
if p_value_ttest < 0.05:
    print("Отвергаем нулевую гипотезу: количество аварий в сентябре статистически значимо больше, чем в других месяцах.")
else:
    print("Не отвергаем нулевую гипотезу: нет статистически значимой разницы в количестве аварий в сентябре и других месяцах.")
```

**Результаты анализа показывают, что количество ДТП в сентябре статистически значимо выше, чем в другие месяцы. Это может свидетельствовать о наличии факторов, характерных именно для этого периода, которые повышают риск ДТП.**

### Гипотеза о тренде снижения смертности (общая)

**Нулевая гипотеза:** Среднее количество смертей от ДТП не изменяется со временем (по годам).

**Альтернативная гипотеза:** Среднее количество смертей от ДТП уменьшается с каждым годом.

**Тест для проверки:**
Использование линейной регрессии для анализа тренда. Линейная регрессия позволяет определить, как целевая переменная (количество смертей) изменяется с годами. Если есть тренд на снижение, коэффициент наклона линии регрессии будет отрицательным. Регрессия позволяет оценить, является ли наблюдаемый тренд статистически значимым с помощью t-статистики и p-значения. Это помогает проверить гипотезу о том, что снижение смертности не случайно, а действительно имеет тенденцию.

```python
# Группировка данных по годам
yearly_deaths = df.groupby('year')['dead_count'].sum().reset_index()

# Линейная регрессия
X = yearly_deaths[['year']]
y = yearly_deaths['dead_count']

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Визуализация
plt.figure(figsize=(10, 5))
sns.scatterplot(data=yearly_deaths, x='year', y='dead_count', label='Фактические данные')
plt.plot(yearly_deaths['year'], y_pred, color='red', label='Линейный тренд')
plt.title('Тренд смертности от ДТП по годам')
plt.xlabel('Год')
plt.ylabel('Суммарное количество смертей')
plt.legend()
plt.show()

# Коэффициент наклона
slope = model.coef_[0]
if slope < 0:
    print(f"Наблюдается статистически значимый тренд на снижение смертности (наклон: {slope:.2f}).")
else:
    print(f"Тренд на снижение смертности отсутствует (наклон: {slope:.2f}).")
```

- частота ДТП по времени суток: Значительное число ДТП приходится в светлое время суток. При этом даже в темное время суток, где освещение отсутствует меньше ДТП, чем там, где освещение включено. В разрезе регионов мы видим, что минимальное значение ДТП в темное время суток с отсутствующим освещением в Москве и Санкт-Петербурге. СКорее всего это связано с инфраструктурой этих городов,т.к. там освещение максимально присутствует во всех частях региона. Судя по этому мы делаем вывод, что в Чеченской республике гораздо больше мест, где освещение отсутствует, поэтому там и больше ДТП в вечернее время. Сумерки занимают довольно маленькую часть на графиках, потому что они вцелом и в сутках длятся всего несколько часов, а в зимние период и того меньше.

- частота ДТП по месяцам: сентябрь может иметь более высокую аварийность из-за начала учебного года и увеличения трафика. 

- изменение смертности в ДТП по годам и регионам: самый низкий процент смертности в Санкт-Петербурге и Москве, самый высокий (и сильно превышающий более чем в два раза относительно других регионов) в Чеченской республике. Во многих регионах замечается тенденция к снижению смертности, что стоит проверить в разделе гипотез, действительно ли эта тенденция статистически значима.

- категории ДТП: Наезд и столкновение являются самыми популярными и равными по количеству категориями при ДТП.

- распределение ДТП по категориям участников: самые распространенные участники ДТП - автомобили (их больше половины), причем по всем регионам преобладает одинаковое распределение между участниками. Судя по графику - самый безопасный способ перемещения - общественный транспорт, даже безопаснее, чем ходить пешком. Пешеходы очень часто попадают в ДТП.

- влияние погоды на количество ДТП: большинство ДТП происходит в ясную хорошую погоду, а также, когда пасмурно. При этом при осадках и в экстремальных условиях ДТП совершается меньше. ХМАО-Югра является северным регионом с продолжительной зимой, поэтому в этом регионе погода значительно влияет на ДТП, также как и Санкт-Петербург - чем больше пасмурных дней, тем выше уровень ДТП, совершенных в пасмурную погоду. При этом в Каснодарском крае, Чеченской республике и Хабаровском крае больше солнечных дней, поэтому ДТП происходят в таких условиях.

- распределение ДТП по степени тяжести: большинство происходящих ДТП заканчиваются с легким исходом, погибших в общей статистике около 8%. Заметна тенденция после 2019 года к снижению ДТП любой степени тяжести.

- влияние категорий участников ДТП на степень тяжести: мотоциклисты пагубно влияют на степень тяжести ДТП, т.к. во всех группах преобладает легкая степень тяжести, а в группе с мотоциклистами - тяжелая.

- влияние плохой погоды на степень тяжести ДТП: погода не влияет на степень тяжести ДТП.

**Выводы по гипотезам:**

**1. Отклонение нулевой гипотезы: Различия в тяжести ДТП между ДТП с участием мотоциклистов и без них статистически значимы. Таким образом, тяжесть ДТП для мотоциклистов отличается от тяжести ДТП для других участников. Мотоциклисты влияют на тяжесть ДТП. Это может быть связано с их большей уязвимостью на дорогах, высокими скоростями, ограниченной защитой при авариях и характером их участия в ДТП.**
